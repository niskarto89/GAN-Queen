```python
# Langkah 1: Import pustaka yang diperlukan
import pandas as pd
import os
import zipfile
from google.colab import files
from reportlab.lib import colors
from reportlab.lib.pagesizes import legal, landscape
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle, Paragraph, Spacer, PageBreak
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.lib.enums import TA_JUSTIFY
from reportlab.lib.units import inch
import uuid  # Untuk menambahkan penanda unik pada nama file

# Langkah 2: Unggah file dari komputer lokal
uploaded = files.upload()

# Langkah 3: Load file Excel yang diunggah
file_path = next(iter(uploaded.keys()))
xls = pd.ExcelFile(file_path)

# Langkah 4: Load sheet yang relevan ke dalam DataFrame
df = pd.read_excel(file_path, sheet_name='Sheet1')

# Langkah 5: Bersihkan dan ekstrak kolom yang diperlukan, lalu hapus kolom 'Bahasa'
df_cleaned = df.iloc[:, [0, 1, 4, 5, 9, 11, 12, 10, 7, 13]]
df_cleaned.columns = [
    'Username', 'Nama Siswa', 'Password', 'Nama Sekolah',
    'Kelas', 'Kategori', 'Sesi', 'Bahasa', 'Nama Pendamping', 'No HP Pendamping'
]
df_cleaned = df_cleaned.drop(columns=['Bahasa'])  # Hapus kolom 'Bahasa'

# Langkah 6: Kelompokkan data berdasarkan 'Nama Pendamping'
grouped_dict = {name: group for name, group in df_cleaned.groupby('Nama Pendamping')}

# Langkah 7: Buat direktori output jika belum ada
output_dir = 'output_data'
os.makedirs(output_dir, exist_ok=True)

# Setup styles untuk header dan footer
styles = getSampleStyleSheet()
header_style = styles['Title']

footer_style = ParagraphStyle(
    'CustomFooter',
    fontSize=12,
    leading=16,
    spaceAfter=10,
    spaceBefore=10,
    alignment=TA_JUSTIFY,
    fontName='Helvetica'
)

section_style = ParagraphStyle(
    'SectionStyle',
    fontSize=14,
    leading=20,
    spaceBefore=15,
    spaceAfter=10,
    fontName='Helvetica-Bold',
    textColor=colors.HexColor('#000066')
)

def create_footer_elements():
    footer_text = []
    footer_text.append(Paragraph("INFORMASI PENTING TANTANGAN BEBRAS 2024", section_style))

    footer_text.append(Paragraph("1. Pra Tantangan", section_style))
    footer_text.append(Paragraph(
        "Setiap peserta diharapkan mencoba mengakses situs tantangan di https://tantanganbebras.ipb.ac.id untuk mencoba akun pada 10-11 November 2024 dan Platform latihan dapat diakses di https://latihan.bebras.or.id (Menggunakan Akun Google) .",
        footer_style
    ))

    footer_text.append(Paragraph("2. Coba Akun/Latihan", section_style))
    footer_text.append(Paragraph(
        "• Diberikan 3 kali kesempatan untuk latihan.<br/>"
        "• Pastikan perangkat dan koneksi internet yang memadai.<br/>"
        "• Pastikan baterai perangkat penuh dan koneksi internet stabil.",
        footer_style
    ))

    footer_text.append(Paragraph("3. Saat Tantangan", section_style))
    footer_text.append(Paragraph(
        "• Ikuti jadwal sesuai kategori dan bahasa.<br/>"
        "• Hanya ada 1 kesempatan.<br/>"
        "• Siapkan kertas dan pensil.<br/>"
        "• Dapat dilakukan secara luring.<br/>"
        "• Laporkan gangguan ke Biro Bebras terkait.",
        footer_style
    ))

    footer_text.append(Paragraph("4. Pengumuman Hasil", section_style))
    footer_text.append(Paragraph(
        "• Hasil diumumkan setelah semua keluhan diproses.<br/>"
        "• Pengumuman di laman http://bebras.or.id.<br/>"
        "• Sertifikat diberikan oleh National Board Organization Bebras.",
        footer_style
    ))

    # Tambahkan PageBreak untuk memulai halaman baru
    footer_text.append(PageBreak())

    # Jadwal Tantangan di halaman baru
    footer_text.append(Paragraph("5. Jadwal Tantangan", section_style))
    footer_text.append(Paragraph(
        "Selasa, 12 November 2024\tPenegak\tTingkat kesulitan kelas 10 – 12<br/>"
        "Rabu, 13 November 2024\tPenggalang\tTingkat kesulitan kelas 7 – 9<br/>"
        "Kamis, 14 November 2024\tSiaga\tTingkat kesulitan kelas 4 – 6<br/>"
        "Jumat, 15 November 2024\tSiKecil\tTingkat kesulitan kelas 1 – 3<br/>"
        "Sabtu, 16 November 2024\tSemua Kategori\tSekolah khusus dengan peserta lebih dari 3000",
        footer_style
    ))

    footer_text.append(Paragraph("Sertifikat", section_style))
    footer_text.append(Paragraph(
        "Dua jenis sertifikat diberikan:<br/>"
        "1. Sertifikat dengan nilai (untuk nilai ≥ 50).<br/>"
        "2. Sertifikat keikutsertaan (untuk nilai < 50).<br/><br/>"
        "Sertifikat dikirim sebagai e-certificate.<br/><br/>"
        "<b>CATATAN PENTING:</b> Guru dimohon teliti dalam pendaftaran untuk menghindari kesalahan sertifikat. Biro dan Panitia tidak bertanggung jawab atas kesalahan setelah pencetakan sertifikat.",
        footer_style
    ))

    footer_text.append(Paragraph(
        "Platform latihan diakses di https://latihan.bebras.or.id berbeda dengan Platform Tantangan yang diakses di https://tantanganbebras.ipb.ac.id dan sengaja dibuat terpisah.<br/>"
        "Walaupun platform nya berbeda, tampilan dan antarmuka interaksi dengan sistem sama sehingga seharusnya peserta yang sudah pernah berlatih di platform latihan akan sudah terbiasa serta tidak canggung berinteraksi dengan sistem.",
        footer_style
    ))

    return footer_text

# Langkah 8: Buat PDF per kelompok
pdf_files = []
for name, group in grouped_dict.items():
    group = group.reset_index(drop=True)
    group.insert(0, 'ID', range(1, len(group) + 1))

    unique_schools = ", ".join(group['Nama Sekolah'].unique())
    unique_id = str(uuid.uuid4())[:8]  # Tambahkan ID unik pendek untuk membedakan file
    file_name = f"{output_dir}/Data_{name.replace(' ', '_').replace(',', '').replace('.', '')}_{unique_schools.replace(' ', '_').replace(',', '').replace('.', '')}_{unique_id}.pdf"

    pdf = SimpleDocTemplate(file_name, pagesize=landscape(legal))
    elements = []

    header_text = f"DAFTAR Peserta Bebras Challenge 2024 - {unique_schools}"
    header = Paragraph(header_text, header_style)
    elements.append(header)
    elements.append(Spacer(1, 12))

    data = [group.columns.tolist()] + group.values.tolist()
    col_widths = [max([len(str(item)) for item in group[col]]) * 6 + 20 for col in group.columns]

    table = Table(data, colWidths=col_widths)
    table.setStyle(TableStyle([
        ('BACKGROUND', (0, 0), (-1, 0), colors.grey),
        ('TEXTCOLOR', (0, 0), (-1, 0), colors.whitesmoke),
        ('ALIGN', (0, 0), (-1, -1), 'LEFT'),
        ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),
        ('BOTTOMPADDING', (0, 0), (-1, 0), 12),
        ('TOPPADDING', (0, 1), (-1, -1), 10),
        ('BOTTOMPADDING', (0, 1), (-1, -1), 10),
        ('GRID', (0, 0), (-1, -1), 1, colors.black),
        ('VALIGN', (0, 0), (-1, -1), 'MIDDLE'),
        ('TEXTCOLOR', (0, 1), (-1, -1), colors.black),
    ]))

    row_colors = [colors.beige, colors.lightgrey]
    for row_num, _ in enumerate(data[1:], start=1):
        table.setStyle(TableStyle([
            ('BACKGROUND', (0, row_num), (-1, row_num), row_colors[row_num % 2])
        ]))

    elements.append(table)
    elements.append(PageBreak())
    footer_elements = create_footer_elements()
    elements.extend(footer_elements)

    pdf.build(elements)
    pdf_files.append(file_name)

# Langkah 9: Buat file ZIP
zip_file_path = f"{output_dir}/Grouped_Student_Data_PDF.zip"
with zipfile.ZipFile(zip_file_path, 'w') as zipf:
    for file in pdf_files:
        base_name = os.path.basename(file)
        counter = 1
        while base_name in zipf.namelist():
            base_name = f"{os.path.splitext(base_name)[0]}_{counter}.pdf"
            counter += 1
        zipf.write(file, arcname=base_name)

# Langkah 10: Download ZIP
files.download(zip_file_path)

print("Proses selesai. File ZIP siap diunduh.")
```
